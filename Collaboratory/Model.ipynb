```python
import os
import random


folder_path = '/content/drive/MyDrive/FinalData/test/real_image'

# Set the probability of an image being removed (e.g., 0.2 for 20% of images)
remove_prob = 0.2

# Get the list of image file names in the folder
file_names = os.listdir(folder_path)

# Loop through the file names and randomly remove images
for file_name in file_names:
    if random.uniform(0, 1) < remove_prob:
        os.remove(os.path.join(folder_path, file_name))
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, concatenate, UpSampling2D, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy,BinaryAccuracy
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import glob
from matplotlib import pyplot as plt
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
%matplotlib inline
```

```python
train_path = '/content/drive/MyDrive/previous/drive/MyDrive/FreshData/train'
test_path = '/content/drive/MyDrive/previous/drive/MyDrive/FreshData/test'
```

```python
#the preprocessig funviotn that is going to apply some preprocesing on the image before passing it to the model,
#we are processing them in the same format as the VGG16 model takes in data
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg19.preprocess_input).flow_from_directory(directory=train_path,target_size=(256,256),classes=['fake_image','real_image'],batch_size=32)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg19.preprocess_input).flow_from_directory(directory=test_path,target_size=(256,256),classes=['fake_image','real_image'],batch_size=32)
# train_batches = ImageDataGenerator(rescale=1./255).flow_from_directory(directory=train_path,target_size=(256,256),classes=['fake_image','real_image'],batch_size=32)
# test_batches = ImageDataGenerator(rescale=1./255).flow_from_directory(directory=test_path,target_size=(256,256),classes=['fake_image','real_image'],batch_size=32)
```

```python
assert train_batches.n == 4500
assert test_batches.n == 1149
assert train_batches.num_classes == test_batches.num_classes == 2
```

```python
imgs,labels = next(train_batches)
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(imgs)
print(labels)
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
```

```python
# define the MesoNet architecture
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.001),loss=BinaryCrossentropy(),metrics=['accuracy'])
```

```python
history = model.fit(train_batches,epochs=28,shuffle=True, verbose=2)
```

```python
predictions = model.predict(test_batches)
print("Predictions:", predictions)
print("Ground Truth Labels:", test_batches.classes)
```

```python
from sklearn.metrics import classification_report
y_pred = np.argmax(predictions, axis=1)

# Get the true class labels
y_true = test_batches.classes

print(classification_report(y_true, y_pred))
```

```python
import pandas as pd
import seaborn as sns

preds_classes = np.argmax(predictions, axis=1)
true_classes = test_batches.classes
class_labels = list(test_batches.class_indices.keys())

# Generating  the confusion matrix
cm = confusion_matrix(true_classes, preds_classes)

cm_df = pd.DataFrame(cm,
                     index = ['fake_image','real_image'], 
                     columns = ['fake_image','real_image'])


plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
print("Confusion Matrix:\n", cm)
```
